# Source generators

In this notebook we have a look at the different source generators that are available.

```python
from auraliser import Auraliser
import numpy as np
#from geometry.core import PointList as PointVector
from geometry import Point
from IPython.lib.display import Audio
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0, 4.0)
%matplotlib inline
import matplotlib.pyplot as plt
import logging
logger = logging.getLogger()
logger.setLevel(logging.INFO)
```

```python
fs = 24000.0      # Sample frequency
duration = 30.      # Duration in seconds
df = 50.0           # Frequency resolution (impedances reflections)

dt = 1.0/fs                             # Seconds per sample
t = np.arange(0.0, duration, dt)        # Time vector

model = Auraliser(duration=duration, sample_frequency=fs)
```

We will need a source

```python
src = model.add_source(name='source', position=Point(0.0,0.0,0.0))
```

```python
subsrc = src.add_subsource(name='subsource')
```

and a receiver.

```python
rcv = model.add_receiver(name='receiver', position=Point(100.0,0.0,0.0))
```

Because we're only interested in the emission we turn off all propagation effects

```python
model.settings['spreading']['include'] = False
model.settings['doppler']['include'] = False
model.settings['atmospheric_absorption']['include'] = False
model.settings['reflections']['include'] = False
model.settings['turbulence']['include'] = False
```

### Sine

Perhaps the most basic generator is the sine generator.

```python
from auraliser.generator import Sine
sine = subsrc.add_virtualsource('sine')
sine.signal = Sine(frequency=100.0)
signal = rcv.auralise()
```

```python
fig = signal.plot_spectrum()
```

```python
fig = signal.plot_leq()
```

```python
model.remove_object('sine')
```

### Noise

```python
from auraliser.generator import Noise
sine = subsrc.add_virtualsource('white')
sine.signal = Noise()
signal = rcv.auralise()
```

```python
fig = signal.plot_spectrum()
```

```python
fig = signal.plot_leq()
```

```python
model.remove_object('white')
```

### White band

```python
from auraliser.generator import WhiteBand
sine = subsrc.add_virtualsource('whiteband')
sine.signal = WhiteBand(1000.0, 2000.0, 50)
signal = rcv.auralise()
```

```python
fig = signal.plot_spectrum()
```

```python
fig = signal.plot_leq()
```

```python
model.remove_object('whiteband')
```
